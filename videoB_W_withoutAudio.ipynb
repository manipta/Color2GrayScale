# **1. Import Libraries**

**1.1 Importing Needed Libraries For Starting**

```python
import cv2
import os
```

# **2. Defining variables**

**2.1 Setting up Directories**

```python
home='/content/drive/MyDrive/ML Classes/ClrVideo2Gray' #Path of Project Folder ------
Extracted_Frames=home+'/ExtractedFrames' 
BW_Frames=home+'/BWFrames'
try: 
    os.mkdir(Extracted_Frames) #Path For All Frames
except OSError as error: 
    print(error) 
try: 
    os.mkdir(BW_Frames) #creating for B/W Frames
except OSError as error: 
    print(error) 
```

**2.2 Defining Variables**

```python
v_path='/content/drive/MyDrive/ML Classes/big_buck_bunny_720p_5mb.mp4' #Path of Colored Video ------
filena=v_path.split("/")[-1]
video=cv2.VideoCapture(v_path)
fps = video.get(cv2.CAP_PROP_FPS) #Get the Frame Rate of Video
bw_path='/content/' #Path for B/W video with name of file ------
bw_video=bw_path+'B&W_'+filena 
```

# **3. Extraction of Frames**

**3.1 Extracting Color Frames From Video**

```python
success = True
count = 1
while success:
  success,frame = video.read()
  name = Extracted_Frames+"/"+str(count)+'.jpg'
  if success == True:
    cv2.imwrite(name,frame)
    print('Frame No. {} Extracted Successfully'.format(count))
    count = count+1
  else:
    break
print('Total {} Frame Extracted Successfully'.format(count-1))
```

# **4. Converting to GrayScale**

**4.1 Converting Colored Extracted Frames to B/W Frames**

```python
import glob
from google.colab.patches import cv2_imshow
cv_img = []
for img in glob.glob(Extracted_Frames+"/*.jpg"):
    n= cv2.imread(img,0)
    cv_img.append(n)
    img=str(img)
    cv2.imwrite(BW_Frames+"/"+img.split("/")[-1],n)
```

# **5. Forming Video From Frames**

**5.1 Forming Videos From B/W Frames**

```python
for img in cv_img:
    height, width = img.shape
    size = (width,height)
out = cv2.VideoWriter(bw_video,cv2.VideoWriter_fourcc(*'MP4V'), fps, size,False)
for i in range(len(cv_img)):
    out.write(cv_img[i])
out.release()
print("Your GrayScale Video is Saved!!")
print("Path: "+bw_video)
```
